Trích xuất dữ liệu thành các phần


```
import pandas as pd
```

```
dataset = pd.read_csv("data2.csv")
test = pd.read_csv("asd.csv")
test
```

```
x = dataset["O(N)"]
x
```

```
y = dataset["Time"]
```

```
import matplotlib.pyplot as plt
plt.plot(x,y,'go')
plt.show()
```

Chia tập dữ liệu thành test/train khác nhau với tỉ lệ 8/2

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(x,y,test_size = 0.01,random_state = 0)
```

```
X_train = X_train.values.reshape(-1,1)
Y_train = Y_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
```

Train dữ liệu và đánh giá mô hình

```
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,Y_train)
model = LR.predict(X_test)
print(LR.score(X_train,Y_train))
```

```
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Y_test,model))
```

---->>  Có thể thấy mô hình được train rất tốtốt

Biểu diễn mô hình trên sơ đồ

```
import matplotlib.pyplot as plt
plt.plot(X_train,Y_train,'.',X_test,model,'-')
plt.show()
```
