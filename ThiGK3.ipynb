Trích xuất dữ liệu thành các phần


```
import pandas as pd
```

```
dataset = pd.read_csv("da.csv")
dataset
```

```
x = dataset["O(N)"]
x
```

```
y = dataset["Time"]
```

Chia tập dữ liệu thành test/train khác nhau với tỉ lệ 8/2

```
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```
X_train = X_train.values.reshape(-1,1)
Y_train = Y_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
```

Train dữ liệu và đánh giá mô hình

```
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,Y_train)
model = LR.predict(X_test)
print(LR.score(X_train,Y_train))
```

```
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Y_test,model))
```
