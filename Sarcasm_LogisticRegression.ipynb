```
df=df.drop('article_link',axis=1)
df
```

Loi b i ct "article_link" do trong qu trnh x l khng cn n.

# Feature Engineering

```
X=df['headline']
y=df['is_sarcastic']
y
```

La chn feature


```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['headline'],df['is_sarcastic'],test_size=0.2,random_state=101)
X_test
```

Chia thnh 2 tp test v train

```
from sklearn.feature_extraction.text import TfidfTransformer,CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

Ci t cc th vin cn thit

```
pipe = Pipeline([('vect',CountVectorizer()),
               ('tfidf',TfidfTransformer()),
               ('model',LogisticRegression())])

model=pipe.fit(X_train, y_train)
```

Thay v lm cc cng on ring, pipeline cho php gp cc cng on li vi nhau.
* Counvectorizer: chuyn text thnh vector
* TFidfTransformer: tnh ton trng s cho ma trn
* LogisticRegression: model la chn  train

```
pred=model.predict(X_test)
print(accuracy_score(y_test,pred))
```

nh gi ca model da trn accuracy

Accuracy l  o c nh gi bng cch (s cu d on ng) / (ton b cu ca vn bn)


```
pred
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```
